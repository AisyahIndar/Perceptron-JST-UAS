```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

url = "https://raw.githubusercontent.com/Kareem-Mohamed-Wardany/Single-Layer-Perceptron/main/penguins.csv"
df = pd.read_csv(url)

df = df[df['species'].isin(['Adelie', 'Gentoo'])]
df['label'] = df['species'].map({'Adelie': 1, 'Gentoo': 0})
df = df[['bill_length_mm', 'flipper_length_mm', 'bill_depth_mm', 'body_mass_g', 'label']].dropna()
df = df.reset_index(drop=True).head(100)

X = df[['bill_length_mm', 'flipper_length_mm', 'bill_depth_mm', 'body_mass_g']].values
y = df['label'].values

learning_rate = 0.1

def step(v):
    return 1 if v > 0 else 0

def train_perceptron(X, y, epochs=5):
    epoch_results = []

    w = np.zeros(4)
    w_lama = np.zeros(4)

    for epoch in range(1, epochs + 1):
        for i in range(len(X)):
            x1, x2, x3, x4 = X[i]
            target = y[i]

            v = np.dot(w, [x1, x2, x3, x4])
            y_pred = step(v)
            error = target - y_pred

            delta_w = w - w_lama

            if error != 0:
                w += learning_rate * error * np.array([x1, x2, x3, x4])

            epoch_results.append({
                "epoch": epoch,
                "x1": x1,
                "x2": x2,
                "x3": x3,
                "x4": x4,
                "V": v,
                "Luaran y'": y_pred,
                "y": target,
                "error": error,
                "w1 baru": w[0],
                "w2 baru": w[1],
                "w3 baru": w[2],
                "w4 baru": w[3],
                "delta w1": w[0] - w_lama[0],
                "delta w2": w[1] - w_lama[1],
                "delta w3": w[2] - w_lama[2],
                "delta w4": w[3] - w_lama[3]
            })

            w_lama = w.copy()

    return pd.DataFrame(epoch_results)

df_result = train_perceptron(X, y, epochs=5)

df_result
```
